3. Compare the Triplets

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'compareTriplets' function below.
#
# The function is expected to return an INTEGER_ARRAY.
# The function accepts following parameters:
#  1. INTEGER_ARRAY a
#  2. INTEGER_ARRAY b
#

def compareTriplets(a, b):
    
    bob = 0
    alice = 0
    n = len(a)
    for i in range(n):
        if a[i]>b[i]:
            alice += 1
        elif b[i]>a[i]:
            bob += 1
            
    return [alice,bob]
        
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    a = list(map(int, input().rstrip().split()))

    b = list(map(int, input().rstrip().split()))

    result = compareTriplets(a, b)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()

```

5. Diagonal Difference

```python
#!/bin/python3

import math
import os
import random
import re
import sys

#
# Complete the 'diagonalDifference' function below.
#
# The function is expected to return an INTEGER.
# The function accepts 2D_INTEGER_ARRAY arr as parameter.
#

def diagonalDifference(arr):
    
    diag1 = 0
    diag2 = 0
    c= 0
    n = len(arr)
    for i in range(n-1,-1,-1):
        diag1 = sum([arr[x][x] for x in range(n)])
        diag2 = sum([arr[x][n - 1 - x] for x in range(n)])
        c += 1
    return abs(diag1-diag2)
if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = []

    for _ in range(n):
        arr.append(list(map(int, input().rstrip().split())))

    result = diagonalDifference(arr)

    fptr.write(str(result) + '\n')

    fptr.close()

```

```python
import math
import os
import random
import re
import sys



def camelcase(s):
    
    count = 1
    for i in s:
        if i.isupper():
            count += 1
    if not s:
        count = 0
    
    return count

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    s = input()

    result = camelcase(s)

    fptr.write(str(result) + '\n')

    fptr.close()
```

```python
import math
import os
import random
import re
import sys



def funnyString(s):
    r = s[::-1]
    n = len(s)
    
    for i in range(1, n):
        if abs(ord(s[i])-ord(s[i-1])) != abs(ord(r[i])-ord(r[i-1])):
            return 'Not Funny'
    return 'Funny'


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        s = input()

        result = funnyString(s)

        fptr.write(result + '\n')

    fptr.close()
```

```python
import math
import os
import random
import re
import sys



def alternatingCharacters(s):
    
    count = 0
    n= len(s)
    for i in range (1,n):
        if s[i]==s[i-1]:
            count += 1
        else:
            count += 0
    return count

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    q = int(input().strip())

    for q_itr in range(q):
        s = input()

        result = alternatingCharacters(s)

        fptr.write(str(result) + '\n')

    fptr.close()
```

```python
def insertion_sort(l):
    for i in range(1, len(l)):
        j = i-1
        key = l[i]
        while (j > 0) and (l[j] > key):
           l[j+1] = l[j]
           j -= 1
        l[j+1] = key

m = int(input().strip())
ar = [int(i) for i in input().strip().split()]
insertion_sort(ar)
print(" ".join(map(str,ar)))
```

5.Running Time of Algorithms

```python
import math
import os
import random
import re
import sys



def runningTime(arr):
    
    count = 0
    n = len(arr)
    for j in range(1,n):
        key = arr[j]
        i = j
        while i >0 and arr[i-1]>key:
            arr[i] = arr[i-1]
            count += 1
            i -= 1
        arr[i] = key
    return count
            

if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = runningTime(arr)

    fptr.write(str(result) + '\n')

    fptr.close()
```

6.Quicksort 1 - Partition

```python
import math
import os
import random
import re
import sys



def quickSort(arr):
    
    left = []
    equal = []
    right = []
    p = arr[0]
    el = 0
    
    for el in arr:
        if el == p:
            equal.append(el)
        elif el < p:
            left.append(el)
        elif el > p:
            right.append(el)
            
    return left + equal + right


if __name__ == '__main__':
    fptr = open(os.environ['OUTPUT_PATH'], 'w')

    n = int(input().strip())

    arr = list(map(int, input().rstrip().split()))

    result = quickSort(arr)

    fptr.write(' '.join(map(str, result)))
    fptr.write('\n')

    fptr.close()
```
